```python
### 멜론 노래 순위 정보 크롤링
from selenium import webdriver
import pandas as pd

path = "C:\\webdriver\chromedriver"
driver = webdriver.Chrome(path)

url = "https://www.melon.com/chart/index.htm"
driver.get(url)
ranks = driver.find_elements_by_css_selector("td:nth-child(2) > div > span.rank")
musics = driver.find_elements_by_css_selector("td:nth-child(6) > div > div > div.ellipsis.rank01 > span > a")
singers = driver.find_elements_by_css_selector("td:nth-child(6) > div > div > div.ellipsis.rank02")
albums = driver.find_elements_by_css_selector("td:nth-child(7) > div > div > div > a")

music = []
for i in musics:
    music.append(i.text)
singer = []
for j in singers:
    singer.append(j.text)
album = []
for k in albums:
    album.append(k.text)
    
```

```python
melon_chart = pd.DataFrame({"제목" : music,
                           "가수" : singer,
                           "엘범" : album},
                           index=[rank.text for rank in ranks])

melon_chart
```

```python
## HTML정보 다운 및 BeautifulSoup을 사용한 처리
from bs4 import BeautifulSoup
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")
songs = soup.select('tr')[1:]
```

```python
### 곡정보 알아오기
song = songs[0]
title = song.select('div.ellipsis.rank01 > span > a')[0].text
title
```

```python
### 가수 찾아오기
singer = song.select('div.ellipsis.rank02 > a')[0].text
singer
```

```python
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    print(title,singer, sep="|")
```

```python
song_data = []
rank = 1

for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    song_data.append(['Melon',rank,title,singer])
    rank += 1
    
song_data
```

```python
### DF로 만들어 사용
import pandas as pd

df_col = ['서비스','순위','타이틀','가수']

pd_df = pd.DataFrame(song_data,columns=df_col)
pd_df.head()
```

```python
### 크롤링 결과를 엑셀로 저장
pd_df.to_excel("data/melon.xlsx",index=False)
```
