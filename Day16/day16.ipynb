```python
jumsu = {
'홍길동' : [50,60,50],
'이순신' : [60,60,60],
'강감찬' : [100,90,90],
'유관순' : [80,80,90],
'강판석' : [95,95,95],
}

## 4개의 Series로 만들어 주세요...
## 이름(name_s) , 각 리스트 0번은 국어(kor_S), 리스트 1번은 영어(eng_S), 리스트 2번은 수학(math_S)
jumsu_ls = list(jumsu.values())
name_S = pd.Series(jumsu.keys())

kor_ls = []
for i in range(len(jumsu_ls)):
    kor_ls.append(jumsu_ls[i][0])


kor_S = pd.Series(kor_ls)

eng_ls = []
for i in range(len(jumsu_ls)):
    eng_ls.append(jumsu_ls[i][1])

eng_S = pd.Series(eng_ls)

math_ls = []
for i in range(len(jumsu_ls)):
    math_ls.append(jumsu_ls[i][2])

math_S = pd.Series(math_ls)
print(math_S)

## DataFrame 만들자...
score = pd.DataFrame({'이름':name_S,'국어':kor_S,'영어':eng_S,'수학':math_S})

### 국어, 영어, 수학 점수 합계 시리즈를 생성하세요...
total = kor_S + eng_S + math_S
print(total)

## 평균값을 구하기
avg = round(total/3,2)
print(avg)

score['합계'] = total
score['평균'] = avg

score
```
