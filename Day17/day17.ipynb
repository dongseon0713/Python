```python
#### 날짜 데이터 생성
dates = pd.date_range('2022-04-01','2022-04-07')
print(dates)
print(type(dates))
## datetime 이라는 특별한 형식 인덱스를 생성함.
## datetime 은 판다스에서 날짜/시간 관련 인덱스. 
```

```python
temps2 = pd.Series([60,70,69,83,77,79,75], index=dates)
print(temps2)
```

```python
# 문제1
### 위의 temps, temps2를 이용하여 뺄셈이외에 다른 연산이 가능한지 실습...
## 더하기, 곱하기, 나누기....
temps_sum = temps1 + temps2
temps_sub = temps1 * temps2
temps_min = temps1 / temps2

# 문제2
### temps3를 생성. temps1과 temps2와 대응하는 시리즈를 생성하세요.
# 온도 : 75, 77, 67, 82, 77, 85, 89 , 인덱스 : dates
#
# 이렇게 생성된 temps, temps2, temps3를 temps_df로 데이터 프레임을 생성하세요.
# temps1 - Missoula, temps2 - Philadelphia, temps3 - Seoul
temps3 = pd.Series([75, 77, 67, 82, 77, 85, 89], index=dates)

temps_df = pd.DataFrame({'Missoula':temps1,'Philadelphia':temps2,'Seoul':temps3})
temps_df

```

```python
## label을 이용한 접근(label은 사용자 정의 인덱스를 말함)
# 접근법
# 1. 단일 : [데이터프레임 or Series].loc['label명',['컬럼명',...]]
# DataFrame
temps_df.loc['20220401']
temps_df.loc['20220401',['Missoula','Seoul']]

# Series
temps_df.Seoul.loc['20220401']

# 2. 여러개일 경우(특정 인덱스만 출력)
temps_df.loc[['20220402','20220406'],['Philadelphia','Seoul']]
# Series
temps_df.Seoul.loc[['20220402','20220406']]

# 3. 여러개일 경우(연속된 값인 경우)
# DataFrame
temps_df.loc['2022-04-02':'20220406',['Philadelphia','Seoul']]
#Series
temps_df.Seoul.loc['20220402':'20220406']
# DataFrame
# 4. Boolean 값을 이용한 출력(이때에 인덱스 값과 같은 계수의 리스트 계수로 생성)
ls = [True,False,False,True,False,False,True]
temps_df.loc[ls,['Philadelphia','Seoul']]
# Series
temps_df.Seoul.loc['20220402':'20220406']
```

```python
## 기본 인덱스를 이용하여 접근
# 접근법
# 1. 단일 : [데이터프레임 or Series].lioc[index값(int)]
# DataFrame
print(temps_df)
temps_df.iloc[3]
temps_df.iloc[3,1]      # Philadelphia의 2022.04.04
# Series
temps_df.Philadelphia.iloc[3]

# 2. 여러개일 경우(특정 인덱스만 출력)
# DataFrame
temps_df.iloc[[1,3,5],[1,2]]
# Series
temps_df.Seoul.iloc[[1,3,5]]

# 3. 여러개일 경우(연속된 값인 경우)
# DataFrame
temps_df.iloc[1:4,[0,2]]

# Series
temps_df.Seoul.iloc[3:]

# 4. Boolean 값을 이용한 출력
# DataFrame
temps_df.iloc[ls,[0,1,2]]

# Series
temps_df.Seoul.iloc[ls]
```
