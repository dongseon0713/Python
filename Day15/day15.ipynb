```python
## 특정 디렉터리 내에 있는 텍스트 자료 수집
import os
 
# 1. 텍스트 디렉터리 경로 지정
print(os.getcwd())  # 현재 기본 경로 확인
txt_data = 'data/txt_data/'

# 2. 텍스트 디렉터리 목록 반환
sub_dir = os.listdir(txt_data)      # first, second
print(sub_dir)

# 3. 각 디렉터리의 텍스트 자료 수집 함수
def textPro(sub_dir):       # first, second
    first_txt = []          # first 디렉터리 텍스트 저장   
    second_txt = []         # second 디렉터리 텍스트 저장
    
    # 디렉터리 구성
    for sdir in sub_dir:
        dirname = txt_data + sdir   # 디렉터리 구성
        file_list = os.listdir(dirname)
        
        # 파일 구성
        for fname in file_list:
            file_path = dirname + '/' + fname
            
            # file 선택
            if os.path.isfile(file_path):
                try:
                    file = open(file_path,'r')
                    if sdir == 'first':
                        first_txt.append(file.read())
                    else:
                        second_txt.append(file.read())
                except Exception as e:
                    print('예외 발생 : ',e)
                finally:
                    file.close()
    return first_txt, second_txt    # 텍스트 자료 반환

### 함수 호출
first_txt , second_txt = textPro(sub_dir)

## 수집한 자료 확인
print('first_txt 길이 : ', len(first_txt))
print('second_txt 길이 : ',len(second_txt))

## 수집한 자료의 결합
tot_texts = first_txt + second_txt
print('tot_texts 길이 : ',len(tot_texts))

## 전체 텍스트 내용
print(tot_texts)
```

```python
### 특수한 파일 처리
# CSV(Comma Seperated Value) : 열단위 자료를 처리할 수 있는 대표적인
# 파일로 CSV, Excel -> Cell단위 자료를 지정하여 처리하는 패턴

# pandas 패키지 import 
import pandas as pd
import os

os.getcwd()

score = pd.read_csv('data/score.csv')
print(score.info())     # 파일 정보
print(score.head())     # 컬럼명 포함 앞 5개 행

# 컬럼 추출
kor = score.kor     # 객체.컬럼명
eng = score['eng']  # 객체['컬럼명']
mat = score['mat']
dept = score.dept

# 각 과목의 최고점 알아오기...
print('국어 최고점 : ',max(kor))
print('영어 최고점 : ',max(eng))
print('수학 최고점 : ',max(mat))

# 각 과목의 최하점 수 알아오기
print('국어 최고점 : ',min(kor))
print('영어 최고점 : ',min(eng))
print('수학 최고점 : ',min(mat))

# 각 과목의 평균 구하기.평균점수는 소수점 2자리까지...
from statistics import mean
print('국어 최고점 : {:.2f}'.format(mean(kor)))
print('영어 최고점 : {:.2f}'.format(mean(eng)))
print('수학 최고점 : {:.2f}'.format(mean(mat)))

count = {}

for cnt in dept:
    count[cnt] = count.get(cnt,0) + 1
    
count
    
```

```python
## Excel 파일 읽기
sam = pd.ExcelFile('data/sam_kospi.xlsx')
print(type(sam))

## excel 파일 파싱
kospi = sam.parse('sam_kospi')
print(kospi.info())

## 컬럼 추출, High, Low
High = kospi['High']            # Serise
Low = kospi['Low']              # Serise

## 최고값(High), 최저값(Low) 평균을 출력
print('최고값 평균 : ',High.mean())
print('최저값 평균 : ',Low.mean())

## 최고값(High), 최저값(Low) 평균을 출력(statistics 모듈)
print('최고값 평균 : ',mean(High))
print('최저값 평균 : ',mean(Low))


```

```python
### JSON 파일
# JSON(JavaScript Object Notation)은 네트워크 표준으로 사용되고 있는
# 파일 양식.(XML...)
#  JSON파일의 특징은 "키-값" 쌍으로 자료를 표현
#   표현은 자바스크립트 구문 형식을 따르지만, 프로그램 언어나 플랫폼
#   에 독립적으로 동작할 수 있음. 
#
# 파이썬은 JSON파일을 처리할 수 있는 모듈을 기본으로 제공하고 있음. 
# 인코딩 - 파이썬 객체(딕셔너리, 리스트)를 -> JSON 문자열("키-값")로 변환
# 디코딩 - JSON 문자열("키-값")을 -> 파이썬 객체(딕셔너리, 리스트)로 변환
# 
# 인코딩 : json.dumps(객체) -> json 문자열로 변경
# 디코딩 : json.loads(json문자열) -> 파이썬 객체로 변경
# #
```

```python
import json

### json 인코딩
user = {'id':1234,'name':'홍길동'}      # 파이썬 딕셔너리
print(type(user))
print(user['name'])

# 제이슨은 무조건 더블쿼터("") 사용

jsonString = json.dumps(user,ensure_ascii=False)
# ensure_ascii 옵션에 False는 아스키코드를 사용하지 않아요...
print(type(jsonString))      # 딕셔너리 형태를 띄지만 JSON의 형태는 문자열이다.
print(jsonString)


pyObj = json.loads(jsonString)
print(type(pyObj))
print(pyObj)
print(pyObj['name'])
```
