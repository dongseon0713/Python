```python
### 연습
# 각 컬럼별 시리즈를 확인하세요
# [입국객수], [국적코드, 입국객수], [국적코드, 성별]
print(sample_1["입국객수"])
print(sample_1[['국적코드','입국객수']])
type(sample_1[['국적코드','입국객수']])
print(sample_1[['국적코드','성별']])
```

```python
### DF에 컬럼 생성.. 기준년월 2019-11
sample_1['기준년월'] = "2019-11"
sample_1

sample_1["test"] = ["2019-11","2019-12","2020-01","2020-02","2020-03","2020-04"]
sample_1

del sample_1["test"]
sample_1
# sample_1 = sample_1.drop(columns='test')
```

```python
### 필터링 하기
# 1. 성별이 남성인 데이터만 출력
man = sample_1['성별'] == '남성'
sample_1[man]

# 2. 입국 객수 150000만을 넘기는 경우 출력
num = sample_1['입국객수'] > 150000
sample_1[num]

# 3. 성별이 남성이면서, 150000만을 넘는 경우
total = (sample_1['성별'] == '남성') & (sample_1['입국객수'] > 150000)
sample_1[total]
```

```python
# 4. 한 컬럼(시리즈)에서 여러 조건을 필터링
# 국적코드가 "A01", "A18" 인 경우만 출력
conditions = (sample_1['국적코드'] == "A01") | (sample_1['국적코드'] == "A18")
sample_1[conditions]

# or

conditions1 = (sample_1['국적코드'].isin(["A01","A18"]))
sample_1[conditions1]

# A01, A18 을 제외한 나머지
sample_1[conditions1==False]
```

```python
## 데이터 합치기
sample_1

# sample_codemaster.xlsx 파일을 code_master(DF)로 불러오세요
code_master = pd.read_excel("data/sample_codemaster.xlsx")
code_master
```

```python
## 옆으로 통합하기
sample_1_code = pd.merge(left=sample_1, 
                         right=code_master,
                         how="left",
                         left_on='국적코드',
                         right_on='국적코드')       # left_on을 사용시에는 right_on도 같이 사용해야한다(공통 부분)
sample_1_code
```

```python
## 옆으로 통합하기2
sample_1_code = pd.merge(left=sample_1, 
                         right=code_master,
                         how="inner",
                         left_on='국적코드',
                         right_on='국적코드')       # left_on을 사용시에는 right_on도 같이 사용해야한다(공통 부분)
sample_1_code
```

```python
### 데이터 통합(아래)
# 1. sample_2 라는 DF을 만드세요. sample_2.xlsx를 사용
sample_2 = pd.read_excel("data/sample_2.xlsx",
                         header=1,
                         skipfooter=2,
                         usecols="A:C")
sample_2['기준년월'] = "2019-12"

# 데이터 통합
sample_2_code = pd.merge(sample_2, code_master,
                         how="left",
                         left_on='국적코드',
                         right_on='국적코드')
sample_2_code
```

```python
### 아래로 통합
sample = sample_1_code.append(sample_2_code,ignore_index=True) # ignore_index=True : 인덱스값을 무시하고 밑으로 이어짐
sample
```

```python
### pandas.concat() 사용하여 병합하기
sample2 = pd.concat([sample_1_code,sample_2_code],
          ignore_index=True)
sample2
```

```python
### 데이터 저장하기(to_excel)
# sample을 데이터 엑셀파일로 저장
sample.to_excel("data/sample.xlsx")
```

```python
## 인덱스 값을 제거하여 저장
sample2.to_excel("data/sample2.xlsx",index=False)
```
